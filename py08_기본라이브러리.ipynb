```python
# 01 ÌÅ¥ÎûòÏä§ ÏÉÅÏÜçÎ∞õÍ≥† Î©îÏÑúÎìú Ï∂îÍ∞ÄÌïòÍ∏∞ 1
class Calculator:
    def __init__(self):
        self.value = 0

    def add(self, val):
        self.value += val

class UpgradeCalculator(Calculator):
    def minus(self, val):
        self.value -= val

cal = UpgradeCalculator()
cal.add(10)
cal.minus(7)

print(cal.value)
```

```python
# 02 ÌÅ¥ÎûòÏä§ ÏÉÅÏÜçÎ∞õÍ≥† Î©îÏÑúÎìú Ï∂îÍ∞ÄÌïòÍ∏∞ 2
class Calculator:
    def __init__(self):
        self.value = 0

    def add(self, val):
        self.value += val

class MaxLimitCalculator(Calculator):
    def add(self, val):
        self.value += val
        if self.value > 100:
            self.value = 100
        
cal = MaxLimitCalculator()
cal.add(50)
cal.add(60)

print(cal.value)
```

```python
from konlpy.tag import Okt

okt = Okt()

positive_words = {
    "ÌñâÎ≥µÌïòÎã§": 2, "ÏÇ¨ÎûëÌïòÎã§": 2, "Í∞êÏÇ¨ÌïòÎã§": 2, "Ï¶êÍ≤ÅÎã§": 2, "ÏïÑÎ¶ÑÎãµÎã§": 2,
    "Í∞êÎèôÌïòÎã§": 1, "ÎßåÏ°±ÌïòÎã§": 1, "Ìù¨ÎßùÌïòÎã§": 2, "Ï°¥Í≤ΩÌïòÎã§": 2, "Í∏∞ÏÅòÎã§": 2,
    "ÎøåÎìØÌïòÎã§": 1, "Í±¥Í∞ïÌïòÎã§": 1, "ÏÑ±Í≥µÌïòÎã§": 2, "ÎØøÏùåÏßÅÌïòÎã§": 1, "Îî∞ÎúªÌïòÎã§": 1,
    "ÏûêÎûëÏä§ÎüΩÎã§": 2, "Ïö©Í∏∞ÏûàÎã§": 1, "Í∏çÏ†ïÏ†ÅÏù¥Îã§": 2, "Î∞ùÎã§": 1, "Ïã†ÎÇòÎã§": 2
}

negative_words = {
    "Ïä¨ÌîÑÎã§": -2, "ÌôîÎÇòÎã§": -2, "ÎØ∏ÏõåÌïòÎã§": -2, "Î∂àÌñâÌïòÎã§": -2, "ÏßÄÍ≤πÎã§": -1,
    "Î¨¥Í∞ÄÏπòÌïòÎã§": -2, "Í¥¥Î°≠Îã§": -2, "Ïã´Îã§": -2, "ÎëêÎ†µÎã§": -1, "Ïö∞Ïö∏ÌïòÎã§": -2,
    "Ï†àÎßùÌïòÎã§": -2, "Î∂àÏïàÌïòÎã§": -1, "ÏßúÏ¶ùÎÇòÎã§": -1, "Ïô∏Î°≠Îã§": -2, "ÌõÑÌöåÌïòÎã§": -1,
    "ÎπÑÏ∞∏ÌïòÎã§": -2, "Ïã§ÎßùÌïòÎã§": -1, "ÏñµÏö∏ÌïòÎã§": -1, "Ï¥àÏ°∞ÌïòÎã§": -1, "Î∂àÌé∏ÌïòÎã§": -1
}

text = input("ÎãπÏã†Ïùò Í∏∞Î∂ÑÏùÑ Ìïú Î¨∏Ïû•ÏúºÎ°ú ÌëúÌòÑÌï¥ Ï£ºÏÑ∏Ïöî: ")
words = okt.morphs(text, stem=True)

score = 0
pos_count = 0
neg_count = 0

for word in words:
    if word in positive_words:
        score += positive_words[word]
        pos_count += 1
    elif word in negative_words:
        score += negative_words[word]
        neg_count += 1

print("ÌòïÌÉúÏÜå Î∂ÑÏÑù Í≤∞Í≥º:", words)
print("Í∞êÏ†ï Ï†êÏàò:", score)

if pos_count > 0 and neg_count > 0:
    print("Í∏çÏ†ïÏ†ÅÏù∏ Í∞êÏ†ïÍ≥º Î∂ÄÏ†ïÏ†ÅÏù∏ Í∞êÏ†ïÏù¥ ÎèôÏãúÏóê ÎäêÍª¥ÏßëÎãàÎã§ üôÇüò¢")
elif score > 0:
    print("Í∏çÏ†ïÏ†ÅÏù∏ Í∞êÏ†ïÏù¥ ÎäêÍª¥ÏßëÎãàÎã§ üòä")
elif score < 0:
    print("Î∂ÄÏ†ïÏ†ÅÏù∏ Í∞êÏ†ïÏù¥ ÎäêÍª¥ÏßëÎãàÎã§ üò¢")
else:
    print("Ï§ëÎ¶ΩÏ†ÅÏù∏ Í∞êÏ†ïÏù¥ÎÑ§Ïöî üòê")
```

```python
import time

def long_task():  # 5Ï¥àÏùò ÏãúÍ∞ÑÏù¥ Í±∏Î¶¨Îäî Ìï®Ïàò
    for i in range(5):
        time.sleep(1) # 1Ï¥àÍ∞Ñ ÎåÄÍ∏∞ÌïúÎã§.
        print("working:%s\n" % i)

print("Start")

for i in range(5):  # long_taskÎ•º 5Ìöå ÏàòÌñâÌïúÎã§.
    long_task()

print("End")
```

```python
# Ïì∞Î†àÎìú ÏÇ¨Ïö©
import time
import threading
start = time.time()
def long_task():  # 5Ï¥àÏùò ÏãúÍ∞ÑÏù¥ Í±∏Î¶¨Îäî Ìï®Ïàò
    data = 0
    for i in range(100 * 1000 * 1000):
        # time.sleep(1) # 1Ï¥àÍ∞Ñ ÎåÄÍ∏∞ÌïúÎã§.
        # print("working:%s\n" % i)
        data += i

print("Start")

threads = [threading.Thread(target = long_task) for i in range(20)]

for t in threads:
    t.start()
for t in threads:
    t.join() # join ÏúºÎ°ú Ïä§Î†àÎìúÍ∞Ä Ï¢ÖÎ£åÎê†ÎïåÍπåÏßÄ Í∏∞Îã§Î¶∞Îã§.

print("End")
print("threading Ïó∞ÏÇ∞ÏãúÍ∞Ñ:", time.time() - start)
```

```python
# Î©ÄÌã∞ÏΩîÏñ¥
import time
import multiprocessing
start = time.time()
def long_task(): # 5Ï¥àÏùò ÏãúÍ∞ÑÏù¥ Í±∏Î¶¨Îäî Ìï®Ïàò
    data = 0
    for i in range(100*1000*1000):
        # time.sleep(1)  # 1Ï¥àÍ∞Ñ ÎåÄÍ∏∞ÌïúÎã§.
        # print("working: %s\n" % i)
        data += i
print("Start")
processes = [ multiprocessing.Process(target = long_task) for i in range(20) ]

for p in processes:
    p.start()
for p in processes:
    p.join()
print("End")
print("processes Ïó∞ÏÇ∞ÏãúÍ∞Ñ:", time.time() - start)
```
