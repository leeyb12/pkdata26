```python
def gugu(n):
    result = []
    i = 1
    while i < 10:
        result.append(n * i)
        i = i + 1
    return result

print(gugu(2))
```

```python
result = 0
for n in range(1, 1000):
    if n % 3 == 0:
        result += n
    elif n % 5 == 0:
        result += n

print(result)
```

```python
def get_total_page(m, n):
    if m % n == 0:
        return m // n
    else:
        return m // n + 1

print(get_total_page(5, 10))
print(get_total_page(15, 10))
print(get_total_page(25, 10))
print(get_total_page(30, 10))
```

```python
import sys as sy

option = sy.argv[1]

if option == '-a':
    memo = sy.argv[2]
    f = open('memo.txt', 'a')
    f.write(memo)
    f.write('\n')
    f.close()
elif option == '-v':
    f= open('memo.txt')
    memo = f.read()
    f.close()
    print(memo)
```

```python
import requests as req

url = "https://api.open-meteo.com/v1/forecast?latitude=35.116741&longitude=129.08540&current_weather=true"
value = req.get(url).json()

tem = value['current_weather']['temperature']
unit = value['current_weather_units']['temperature']

print(f'현재 온도: {tem}{unit}')
```
