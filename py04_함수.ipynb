## 여러 개의 입력값을 받는 함수 만들기

```python
def add_mul(choice, *args):
    if choice == "add":  # 매개변수 choice에 "add"를 입력받았을 때
        result = 0
        for i in args:
            result = result + i
    elif choice == "mul":  # 매개변수 choice에 "mul"을 입력받았을 때
        result = 1
        for i in args:
            result = result * i
    return result
```

```python
result = add_mul('add', 1, 2, 3, 4, 5)
print(result)
result = add_mul('mul', 1, 2, 3, 4, 5)
print(result)
```

```python
# 위 예제에서 args를 하나만 넣고도 동작하도록 만들어보기
def add_mul(choice, *args):
    if choice == "add":
        # 인자가 없으면 0 반환
        result = sum(args) if args else 0
    elif choice == "mul":
        # 인자가 없으면 1 반환
        result = 1
        for i in args:
            result *= i
    return result
```

```python
print(add_mul("add", 3))        # 3
print(add_mul("add", 1, 2, 3))  # 6
print(add_mul("add"))           # 0

print(add_mul("mul", 5))        # 5
print(add_mul("mul", 2, 3, 4))  # 24
print(add_mul("mul"))           # 1
```

## 함수의 반환값은 언제나 하나이다

```python
def add_and_mul(a, b):
    return a + b, a * b
    # return [a + b, a * b]
    return {a + b : a * b}
```

```python
value = add_and_mul(3, 5)
print(value, type(value))
```

## 매개변수에 초깃값 미리 설정하기 * 

```python
# 매개변수에 초깃값 미리 설정은 함수를 생성할 때 만든다.
def say_myself(name, age, man=True):
    print("나의 이름은 %s입니다." % name)
    print("나이는 %d살입니다." % age)
    if man:
        print("남자입니다.")
    else:
        print("여자입니다.")
say_myself("홍길동", 22)
say_myself("박서방", 23, man=True)
say_myself("향단이", 24, man=False)
```

```python
# json 파싱하여 lambda로 키를 설정하여 정렬하기
import requests as req
url = "https://jsonplaceholder.typicode.com/todos"
result = req.get(url).json()
for item in result:
    print(item)
```

```python
# 1. id의 번호를 기준으로 내림차순으로 정렬하라
import requests as req
url = "https://jsonplaceholder.typicode.com/todos"
result = req.get(url).json()

sorted_result = sorted(result, key=lambda x: x["id"], reverse=True)

for item in sorted_result:
    print(item)
```

```python
# 2. completed가 True된 내용만 전처리하라.
import requests as req
url = "https://jsonplaceholder.typicode.com/todos"
result = req.get(url).json()

sorted_result2 = sorted([x for x in result if x["completed"] == True], key=lambda x: x["id"], reverse=True)

for item2 in sorted_result2:
    print(item2)
```

```python
# 3. 리스트의 각요소 끝에 score 이름으로 키를 만들고 그 키의 값을 40 ~ 100까지의 임의 값으로 배정하여 fix라는 변수에 등록하라.
import requests as req
import random

url = "https://jsonplaceholder.typicode.com/todos"
result = req.get(url).json()

for item in result:
    item["score"] = random.randint(40, 100)

fix = result

for item in fix:
    print(item)
```

```python
# 4. score를 키로 내림차순으로 sorted하라
import requests as req
url = "https://jsonplaceholder.typicode.com/todos"
result = req.get(url).json()
sorted_fix = sorted(fix, key=lambda x: x["score"], reverse=True)

for item in sorted_fix:
    print(item)
```

```python
import requests as req
url = "https://finance.naver.com/sise/sise_market_sum.naver"
ori = req.get(url).text

code = ['005930', '000270', '005380', '012450', '068270']

def finance(code = '005930'):
    start_title = ori.find(f'<a href="/item/main.naver?code={code}" class="tltle">')
    ori1 = ori[start_title : start_title + 70].replace(f'<a href="/item/main.naver?code={code}" class="tltle">','').replace('</a></td>','').replace('</a></td','').replace('\n','').strip()
    ori2 = ori[start_title + 70 : start_title + 105].replace('<td class="number">','').replace('</td>','').replace('>','').strip()
    print("종목:",ori1)
    print("가격:",ori2)

for x in code:
    finance(x)
```
