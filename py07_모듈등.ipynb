```python
# except는 뒤에 나오는 구문을 생략할 수 있다.
try:
    4 / 0
except ZeroDivisionError as e:
    print(e)
```

```python
# finally는 무조건 실행
try:
    f = open('foo.txt', 'w')
    # 무언가를 수행한다.
finally:
    f.close() # 중간에 오류가 발생하더라도 무조건 실행된다.
    print("파일을 열고 닫았습니다.")
```

```python
# 먼저 만나는 오류를 출력
try:
    a = [1, 2]
    print(a[3])
    4 / 0
except ZeroDivisionError as e:
    print("0으로 나눌 수 없습니다.", e)
except IndexError as e:
    print("인덱싱 할 수 없습니다.", e)
```

```python
# 먼저 만나는 오류를 출력
try:
    a = [1, 2]
    print(a[3])
    4 / 0
except (ZeroDivisionError, IndexError) as e:
    print("인덱싱 할 수 없습니다.", e)
```

```python
# try-else 문
try:
    age=input('나이를 입력하세요: ')
except:
    print('입력이 정확하지 않습니다.')
else:
    try:
        if age <= 18:
            print('미성년자는 출입금지입니다.')
        else:
            print('환영합니다.')
    except:
        print("프로그램 점검요망.")
```

```python
# 오류 회피하기
students = ['김철수', '이영희', '박민수', '최유진']

for (i, student) in enumerate(students, 1):
    try:
        with open(f"{student}_성적.txt",'r') as f:
            score = f.read()
            print(f"{student}의 성적: {score}")
    except:
        print(f"{i}: {student}의 성적 파일이 없습니다. 건너뜁니다.")
        continue
```

```python
class MyError(Exception):
    def __str__(self): # 프린트 전용 구문
        return "나는 바보가 아냐! 코훌쩍..."
```

```python
def say_nick(nick):
    if nick == '바보':
        raise MyError()
    print(nick)
```

```python
try:
    print(say_nick("천사"))
    print(say_nick("바보"))
except MyError as e:
    print(e)
```

```python
# p.295
# 01 클래스 상속받고 메서드 추가하기 1
class Calculator:
    def __init__(self):
        self.value = 0
    def add(self, val):
        self.value += val

class UpgradeCalculator(Calculator):
    def minus(self, val):
        self.value -= val

cal = UpgradeCalculator()
cal.add(10)
cal.minus(7)

print(cal.value)
```

```python
# 02 클래스 상속받고 메서드 추가하기 2
class Calculator:
    def __init__(self):
        self.value = 0

    def add(self, val):
        self.value += val

class MaxLimitCalculator(Calculator):
    def add(self, val):
        self.value += val
        if self.value > 100:
            self.value = 100
        
cal = MaxLimitCalculator()
cal.add(50)
cal.add(60)

print(cal.value)
```
