```python
# 응용 

import requests as req
from bs4 import BeautifulSoup as bs
url = "https://www.melon.com/chart/index.htm"
headers = {"user-agent":"Mozilla/5.0 (Window NT 10.0: Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/145.0.0.0 Safari/537.36 Edg/145.0.0.0"}
web = req.get(url, headers = headers)
soup = bs(web.content, 'html.parser')

title = soup.select('.rank01')[:100]
name = soup.select('.rank02 > span')[:100]
album = soup.select('.rank03 > a')[:100]
str = ''
for i, (t, n, a) in enumerate(zip(title, name, album), 1):
    str += f'{i}위: {t.text.strip()} : {n.text.strip()} / 앨범: {a.text.strip()}\n'

print(str)
```

```python
import requests as req
from bs4 import BeautifulSoup as bs
url = "https://www.pusan.ac.kr/kor/CMS/MenuMgr/menuListOnBuilding.do?mCode=MN202"
headers = {"user-agent":"Mozilla/5.0 (Window NT 10.0: Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/145.0.0.0 Safari/537.36 Edg/145.0.0.0"}
web = req.get(url, headers = headers)
soup = bs(web.content, 'html.parser')
menulist = soup.select('.menu-tbl')
menu = soup.select('h3.menu-tit01 + p')
won = soup.select('.menu-tit01')
day = soup.select('.day')
date = soup.select('.date')
for d, dd, w, m in zip(day, date, won, menu):
    print('=' * 15)
    print(d.text.strip(), dd.text.strip())
    print('-' * 15)
    print(w.text + '\n' + m.text.replace('\r', '\n'))
```
