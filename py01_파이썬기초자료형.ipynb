```python
# 문자열
print("%s" % '12.34 문자표현')
# 정수
print("%d" % 12.34)
# 부동소수점
print("%f" % 12.34)
# %%표기
print("한국에 산은 %d%% 정도 있습니다." % 70)
```

```python
# 문자열
print("%-20s입니다" % '12.34 문자표현')
# 정수
print("%-10d" % 12.34)
# 부동소수점
print("%f" % 12.34)
# %%표기
print("한국에 산은 %d%% 정도 있습니다." % 70)
```

### format 함수를 사용한 포매팅

```python
# 왼쪽 정렬
print("{0:<10}".format("hi"))
# 오른쪽 정렬
print("{0:>10}".format("hi"))
# 가운데 정렬
print("{0:^10}".format("hi"))
```

```python
# 응용
re = "{0:.^18}".format("영수증")
print("=" * 20 + "\n" + re + "\n" + "=" * 20)
```

```python
# input으로 개수를 입력받아 한잔에 2500원인 아이스아메리카노와 3800원인 라떼를 계산한 영수증을 출력하라.
```

```python
price1 = 2500
price2 = 3800
count1 = int(input("아아 개수를 입력하세요: "))
count2 = int(input("라떼 개수를 입력하세요:"))
total1 = price1 * count1
total2 = price2 * count2
total3 = total1 + total2
re = "{0:^18}".format("영수증")
print("=" * 20 + "\n" + re + "\n" + "=" * 20)
print("    아아:", count1, "잔    ") 
print("    라떼:", count2, "잔    ")
print("--------------------")
print("    아아:", total1,"원"   )
print("    라떼:", total2,"원"   )
print("--------------------")
print(">>> 총 금액:", total3, "원") 
print("=" * 20)
```

```python
# upper - 소문자를 대문자로 바꾸기
a = 'Python is the best choice'
print(a.upper())
# lower - 대문자를 소문자로 바꾸기
b = 'Python Is The Best Choice'
print(b.lower())
c = 'Python Is The Best Choice'
print(c[0] + c[1:].lower())
```

```python
ss = f"{'hi':^12}"
print(ss)
# lstrip - 왼쪽공백지우기
print(ss.lstrip())
# rstrip - 오른쪽공백지우기
print(ss.rstrip())
# strip - 양쪽공백지우기
print(ss.strip())
```

```python
# replace - 문자열 바꾸기 **
a = "Life is too short"
print(a.replace("too short", "Egg"))
print(a)
```

```python
# split = 문자열 나누기
a = "Life is too short"
print(a.split())
b = "a:b:c:d:e:f"
print(b.split(':'))
```

```python
c = "BaNaNa"
# BaNaNa => banana -> b*a*n*a*n*a -> ['b','a','n','a','n','a']
c1 = c.lower()
c2 = "*".join(c1)
c3 = c2.split('*')
print(c)
print(c1)
print(c2)
print(c3)
```

```python
# isalpha - 문자열이 알파벳으로만 구성되어 있는지 확인하기
s = "Life is too short"
print(s.replace(" ", "").isalpha())  # 메소드체이닝
```

```python
# isdigit - 문자열이 숫자로만 구성되어 있는지 확인하기
n = "123456"
print(n.isdigit())
```

```python
# startswith - 문자열이 특정 문자(열)로 시작하는지 확인하기
s = "Life is too short"
print(s.startswith("Life"))
```

```python
# endswith - 문자열이 특정 문자(열)로 끝나는지 확인하기
s = "Life is too short"
print(s.endswith("short"))
```
